```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from google.colab import drive

drive.mount('/content/drive')

DATA_DIR = "/content/drive/MyDrive/chi_tay"

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

train_generator = datagen.flow_from_directory(
    DATA_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='training'
)

val_generator = datagen.flow_from_directory(
    DATA_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', subset='validation'
)

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(256, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(train_generator.num_classes, activation='softmax')
])

model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_generator, validation_data=val_generator, epochs=30)

model.save('/content/drive/MyDrive/model_chi_tay.h5')
print("ДҗГЈ hoГ n tбәҘt huбәҘn luyб»Үn vГ  lЖ°u mГҙ hГ¬nh chб»ү tay!")
```

### LГҪ do phбә§n trДғm chГӯnh xГЎc (Accuracy) cao bбәҘt thЖ°б»қng vГ  cГЎch khбәҜc phб»Ҙc

Phбә§n trДғm chГӯnh xГЎc cao, Д‘бә·c biб»Үt lГ  `val_accuracy: 1.0000` (100%), lГ  mб»ҷt dбәҘu hiб»Үu cбәЈnh bГЎo rбәұng mГҙ hГ¬nh cб»§a bбәЎn khГҙng Д‘ГЎng tin cбәӯy. DЖ°б»ӣi Д‘Гўy lГ  nhб»Ҝng lГҪ do chГӯnh:

1.  **Tбәӯp dб»Ҝ liб»Үu kiб»ғm Д‘б»Ӣnh (Validation Set) quГЎ nhб»Ҹ**: TrГ¬nh tбәЎo dб»Ҝ liб»Үu cб»§a bбәЎn (`val_generator`) chб»ү tГ¬m thбәҘy **1 бәЈnh** duy nhбәҘt cho tбәӯp kiб»ғm Д‘б»Ӣnh. ДҗбәЎt Д‘Ж°б»Јc 100% Д‘б»ҷ chГӯnh xГЎc trГӘn mб»ҷt Д‘iб»ғm dб»Ҝ liб»Үu duy nhбәҘt hoГ n toГ n khГҙng cГі ГҪ nghД©a vГ  khГҙng thб»ғ hiб»Үn khбәЈ nДғng tб»•ng quГЎt hГіa cб»§a mГҙ hГ¬nh. Kбәҝt quбәЈ nГ y cГі thб»ғ do ngбә«u nhiГӘn hoбә·c mГҙ hГ¬nh chб»ү Д‘ЖЎn giбәЈn lГ  "ghi nhб»ӣ" Д‘Ж°б»Јc бәЈnh Д‘Гі.
2.  **Tбәӯp dб»Ҝ liб»Үu huбәҘn luyб»Үn (Training Set) rбәҘt nhб»Ҹ**: `train_generator` chб»ү tГ¬m thбәҘy **11 бәЈnh** Д‘б»ғ huбәҘn luyб»Үn. CГЎc mГҙ hГ¬nh hб»Қc sГўu (deep learning) cбә§n mб»ҷt lЖ°б»Јng lб»ӣn dб»Ҝ liб»Үu (hГ ng trДғm Д‘бәҝn hГ ng nghГ¬n бәЈnh cho mб»—i lб»ӣp) Д‘б»ғ hб»Қc cГЎc Д‘бә·c trЖ°ng cГі ГҪ nghД©a vГ  hoбәЎt Д‘б»ҷng tб»‘t trГӘn dб»Ҝ liб»Үu mб»ӣi. Vб»ӣi quГЎ Гӯt бәЈnh, mГҙ hГ¬nh rбәҘt dб»… bб»Ӣ **quГЎ khб»ӣp (overfitting)**, nghД©a lГ  nГі chб»ү ghi nhб»ӣ dб»Ҝ liб»Үu huбәҘn luyб»Үn mГ  khГҙng hб»Қc Д‘Ж°б»Јc cГЎc quy tбәҜc cЖЎ bбәЈn.
3.  **Дҗб»ҷ chГӯnh xГЎc khГҙng б»•n Д‘б»Ӣnh**: BбәЎn cГі thб»ғ thбәҘy `val_accuracy` nhбәЈy vб»Қt giб»Ҝa `1.0000` vГ  `0.0000e+00`. Sб»ұ khГҙng б»•n Д‘б»Ӣnh nГ y lГ  mб»ҷt dбәҘu hiб»Үu khГЎc cб»§a viб»Үc Д‘ГЎnh giГЎ khГҙng Д‘ГЎng tin cбәӯy do dб»Ҝ liб»Үu khГҙng Д‘б»§.

### Дҗб»ғ cбәЈi thiб»Үn mГҙ hГ¬nh vГ  cГі Д‘б»ҷ chГӯnh xГЎc Д‘ГЎng tin cбәӯy hЖЎn, bбәЎn nГӘn thб»ұc hiб»Үn cГЎc bЖ°б»ӣc sau:

1.  **TДғng kГӯch thЖ°б»ӣc tбәӯp dб»Ҝ liб»Үu (Quan trб»Қng nhбәҘt)**: Thu thбәӯp thГӘm Д‘ГЎng kб»ғ sб»‘ lЖ°б»Јng бәЈnh cho mб»—i lб»ӣp. ДҗГўy lГ  cГЎch hiб»Үu quбәЈ nhбәҘt Д‘б»ғ cбәЈi thiб»Үn hiб»Үu suбәҘt vГ  Д‘б»ҷ tin cбәӯy cб»§a mГҙ hГ¬nh hб»Қc sГўu. HГЈy cб»‘ gбәҜng cГі Гӯt nhбәҘt 50-100 бәЈnh cho mб»—i lб»ӣp, hoбә·c lГҪ tЖ°б»ҹng lГ  vГ i trДғm бәЈnh trб»ҹ lГӘn.
2.  **TДғng cЖ°б»қng dб»Ҝ liб»Үu (Data Augmentation) mбәЎnh mбәҪ hЖЎn**: Mбә·c dГ№ bбәЎn Д‘ang sб»ӯ dб»Ҙng tДғng cЖ°б»қng dб»Ҝ liб»Үu, vб»ӣi mб»ҷt tбәӯp dб»Ҝ liб»Үu nhб»Ҹ nhЖ° vбәӯy, bбәЎn cГі thб»ғ cбә§n ГЎp dб»Ҙng cГЎc phГ©p biбәҝn Д‘б»•i mбәЎnh mбәҪ hЖЎn Д‘б»ғ tДғng cЖ°б»қng nhГўn tбәЎo kГӯch thЖ°б»ӣc vГ  sб»ұ Д‘a dбәЎng cб»§a dб»Ҝ liб»Үu huбәҘn luyб»Үn. Дҗiб»Ғu nГ y giГәp mГҙ hГ¬nh nhГ¬n thбәҘy nhiб»Ғu biбәҝn thб»ғ hЖЎn tб»« cГЎc бәЈnh hiб»Үn cГі cб»§a bбәЎn. TГҙi sбәҪ Д‘iб»Ғu chб»үnh `ImageDataGenerator` cб»§a bбәЎn Д‘б»ғ lГ m Д‘iб»Ғu nГ y.
3.  **Hб»Қc chuyб»ғn giao (Transfer Learning)**: Sб»ӯ dб»Ҙng mб»ҷt mГҙ hГ¬nh Д‘ГЈ Д‘Ж°б»Јc huбәҘn luyб»Үn sбәөn (nhЖ° VGG, ResNet, MobileNet) trГӘn mб»ҷt tбәӯp dб»Ҝ liб»Үu rбәҘt lб»ӣn (vГӯ dб»Ҙ: ImageNet). BбәЎn cГі thб»ғ sб»ӯ dб»Ҙng mГҙ hГ¬nh nГ y nhЖ° mб»ҷt bб»ҷ trГӯch xuбәҘt Д‘бә·c trЖ°ng hoбә·c tinh chб»үnh (fine-tune) nГі trГӘn tбәӯp dб»Ҝ liб»Үu nhб»Ҹ cб»§a bбәЎn. Дҗiб»Ғu nГ y cho phГ©p mГҙ hГ¬nh cб»§a bбәЎn tбәӯn dб»Ҙng kiбәҝn thб»©c Д‘ГЈ hб»Қc tб»« hГ ng triб»Үu бәЈnh.
4.  **Дҗб»•i tГӘn tб»Үp (cho lб»—i `FileNotFoundError`)**: Mбә·c dГ№ `os.walk` Д‘ГЈ liб»Үt kГӘ Д‘Ж°б»Јc tб»Үp, nбәҝu lб»—i `FileNotFoundError` vбә«n tiбәҝp diб»…n, Д‘iб»Ғu nГ y thЖ°б»қng lГ  do cГЎc kГҪ tб»ұ Д‘бә·c biб»Үt (`taМүi xuoМӮМҒng`) trong tГӘn tб»Үp gГўy ra sб»ұ cб»‘ vб»ӣi cГЎch cГЎc thao tГЎc hб»Ү thб»‘ng tб»Үp cб»§a `ImageDataGenerator` xб»ӯ lГҪ Unicode. HГЈy cГўn nhбәҜc Д‘б»•i tГӘn cГЎc tб»Үp vГ  thЖ° mб»Ҙc cб»§a bбәЎn chб»ү sб»ӯ dб»Ҙng cГЎc kГҪ tб»ұ ASCII cЖЎ bбәЈn (vГӯ dб»Ҙ: `tai_xuong_26.jpg`) Д‘б»ғ Д‘бәЈm bбәЈo khбәЈ nДғng tЖ°ЖЎng thГӯch.
